```python
import json
import psycopg2
import os

# DB 접속 정보 설정
# .env 파일 혹은 시스템 환경변수에 설정된 값을 사용합니다.
CONNECTION_STRING = os.getenv("POSTGRES_CONNECTION_STRING")

def insert_json_to_db(file_path):
    try:
        # JSON 파일 읽기
        with open('C:\\big20\\Final_Project\\big20_AI_Interview_simulation\\Data\\data.json', 'r', encoding='utf-8') as f:
            data_list = json.load(f)

        # 데이터베이스 연결
        conn = psycopg2.connect(CONNECTION_STRING)
        cur = conn.cursor()

        # 데이터 삽입 실행
        for item in data_list:
            # json의 키값이 실제 파일과 일치해야 합니다.
            sql = "INSERT INTO interview_questions (id, question, answer) VALUES (%s, %s, %s)"
            cur.execute(sql, (item['id'], item['question'], item['answer']))

        # 최종 저장 및 종료
        conn.commit()
        print(f"성공적으로 {len(data_list)}개의 데이터를 넣었습니다!")

    except Exception as e:
        print(f"에러 발생: {e}")
        if 'conn' in locals():
            conn.rollback()
    finally:
        if 'cur' in locals(): cur.close()
        if 'conn' in locals(): conn.close()

# 실행
insert_json_to_db('data.json')
```
