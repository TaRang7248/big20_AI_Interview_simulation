#### selenium 정리
1. 브라우저 제어 
    - 탐색(메소드)
        - get(url:str) : 지정된 url 이동, 페이지가 완전히 로드될 때까지 기다려야 한다. (wait)
        - back() : 뒤로가기 버튼 클릭
        - forward() : 앞으로가기
        - refresh() : 새로고침
    - 탐색(속성)
        - current_url : 현재 페이지의 URL 주소 가져오기, 속성임

2. 웹 요소 탐색
    - 단일 요소 찾기
        - driver.find_element(by:By, value:str) : 첫번째 1개 반환
            - By.ID, By.CLASS_NAME, By.TAG_NAME, By.CSS_SELECTOR, By.XPATH
    - 여러 요소 찾기
        - driver.find_elements(by:By, value:str) : 리스트반환
3. 웹 요소 조작 및 정보 추출 : 웹 요소 객체를 선택하고 조작한다
    - 메소드
        - element.click() : 클릭가능
        - element.send_keys(*value) : 선택한 요소(text, select)
        - element.clear() : value 삭제 
        - element.submit(): 제출 
        - element.get_attribute(name) : 태그에 있는 속성의 값을 반환
    - 속성
        - element.text() : 화면에 표시된 텍스트 반환
4. 대기(Waits)
    - 모든 페이지는 즉시 로드되지 않는다. Ajax(비동기 통신), selenium에서 특정 요소를 찾기를 하면 에러발생
        반드시 대기를 해야함

    - 명시적 대기 - 권장
        특정 조건(Expected Condition, EC) : 충족될 때까지 지정한 최대 시간동안 기다리는 방식, 안전
        사용법:
            - from selenium.webdriver.support.ui import WebDriverWait
            - from selenium.webdriver.support import expected_conditions as EC
                - 주요 조건
                    - EC.visibility_of_element_located(By, value) : 특정 요소가 화면에 보일 때까지
                    - EC.presence_of_element_located((By,value)) : 특정 요소가 DOM에 존재할 때까지
                    - EC.element_to_be_clickable((By,value)) : 특정 요소가 클릭 가능할 때까지
    - 묵시적 대기 - 조건 없이 특정 시간만큼 무조건 대기

```python
# 기본 설정 및 브라우저 실행
from selenium import webdriver      #브라우저별 제어방식, 크롬드라이버, 엣지드라이버, 사파리? ,
from selenium.webdriver.chrome.service import Service as ChromeService      #크롬브라우저 제어
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys     # Enter, Tab

# wait 예시
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
options = Options()
service = ChromeService()
driver = webdriver.Chrome(service=service,options=options) # 브라우저가 화면 표시
# driver.het('https://www.naver.com)
# 1. WebDriverWait 객체 생성
wait = WebDriverWait(driver, 10) # driver를 대상으로 최대 10초까지 대기시키겠다

#. until()메소드로 조건이 충족될 때까지 대기
try: 
    element = wait.until(               # 화면에 보일 때까지 대기이잇
        EC.visibility_of_element_located(By.ID,'login')
    )
    element.click()
except TimeoutError as e:
    print(e)
```

```python
import pandas as pd
import numpy as np
from glob import glob # 디렉토리에 특정파일명을 검색, 파일목록을 리스트로 반환
```

```python
stations_filename=glob('../data/지역_*.xls')

tmp_df = pd.read_excel(stations_filename[0],header=2)
tmp_df.head()
```

```python
tmp_row = []
for filename in stations_filename:
    tmp = pd.read_excel(filename, header=2)
    tmp_row.append(tmp)
    
station_raw = pd.concat(tmp_row)
station_raw.info()
```

```python
stations = pd.DataFrame({
    '상호' : station_raw['상호'],
    '주소' : station_raw['주소'],
    '휘발유' : station_raw['휘발유'],
    '셀프여부' : station_raw['셀프여부'],
    '상표' : station_raw['상표'],
})
stations.head()

```

```python
stations['구'] = [addr.split()[1] for addr in stations['주소']]
stations['구'].nunique()
```

```python
stations[stations['휘발유'] =='-']
```

```python
stations.reset_index(inplace=True)
del stations['index']
stations.head()
```
