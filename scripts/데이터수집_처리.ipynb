```python
import pandas as pd
import numpy as np
import requests
```

```python
url = 'https://www.google.com'
response = requests.get(url)

if response.status_code==200:
    print(response.text[:100])
else:
    print(f'Request Failed: {response.status_code}')
```

```python
# 예외 처리 방식 권장
url = 'https://www.google.com'

while(True):
    try:
        response = requests.get(url,timeout=3)
        
        # 400,500 에러 발생시 예외 발생 
        response.raise_for_status()
        # 데이터 처리
        print(response.text[:100])
        break
    except requests.exceptions.HTTPError as errh:       # 프로토콜 에러
        print(f'HTTP Error : {errh}')
    except requests.exceptions.ConnectionError as errc:
        print(f'Connection Error : {errc}')
    except requests.exceptions.Timeout as errt:
        print(f'TimeOut Error : {errt}')
    except requests.exceptions.RequestException as errreq:
        print(f'Somthing Error{errreq}')
        
```

```python
# form dasta 전송 post test
url = 'https://httpbin.org/post'
form_data = {
    'username' : 'myname',
    'password' : '1234',
}
response = requests.post(url,data=form_data)
print(response.json())
```

```python
# json data 전송 post test
url = 'https://httpbin.org/post'

# 전송할 딕셔너리 데이터
dict_data = {
    'custid':'1000',
    'prodid':'200',
    'tag' : ['elec','computer','']
}

response = requests.post(url,json=dict_data)
print(response.json())
print(response.json()['json'])
```
